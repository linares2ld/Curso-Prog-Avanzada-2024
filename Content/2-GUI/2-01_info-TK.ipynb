```python
import tkinter as tk
from tkinter import font
```

__Versión de Tkinter:__

```python
# Utiliza la clase Tcl para ejecutar un comando Tcl (info patchlevel) que devuelve la versión de Tkinter

print(f"Versión de Tkinter: {tk.Tcl().eval('info patchlevel')}\n")
```

__Ventana principal:__

```python
# Crea la ventana principal de la aplicación Tkinter y la asigna a la variable root

root = tk.Tk()
```

# Fuentes disponibles

```python
# - Obtiene una lista de todas las familias de fuentes disponibles
# - Almacena la información en la variable fonts

fonts = list(font.families())

for i in fonts:
    print(i)
print("")
```

```python
# Muestra las caracteristicas de la fuente por defecto actual:

default_font = font.nametofont("TkDefaultFont")
print(f"{default_font.actual()}\n")

# Muestra las caracteristicas de la nueva fuente por defecto:

default_font.configure(family="Times New Roman", size=12)
print(f"{default_font.actual()}\n")
```

# Mostrar ventana principal

```python
# Establece el título de la ventana principal
root.title('Font Families')

# Crea un widget de etiqueta con la fuente por defecto
# - Como se modifico la fuente por defecto en la celda anterior, se mostrará con fuente modificada y no la original
etiqueta = tk.Label(root, text="Hola, mundo!")

# Muestra la etiqueta en la ventana
etiqueta.pack()

# Inicia el bucle principal
# Es necesario para que la ventana se mantenga visible
root.mainloop()
```
